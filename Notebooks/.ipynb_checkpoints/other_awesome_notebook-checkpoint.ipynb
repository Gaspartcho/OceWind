```julia

function plot_data(data, title)
    fig = Figure(size = (400, 300))
    
    ax = Axis(fig[1, 1]; title=title, xlabel = "t (minutes)", ylabel = "z (meters)",)

    dims = size(data)
    center_x = 1:dims[1]
    center_y = 1:dims[2]
    
    hm_w = heatmap!(ax, center_x, center_y, data)
    Colorbar(fig[:, end+1], hm_w)

    return fig


end
```
