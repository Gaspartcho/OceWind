```julia
filepath = "/home/gaspard/Documents/Ecole/BX S2/Stage/OceWind/results/data/ocean_wind_mixing_and_convection_rank"
nb_ranks = 4
```

```julia
@info "Importing Data..."
	
filename = string(filepath, 0, ".nc")



u = ncread(filename, "u")
v = ncread(filename, "v")
w = ncread(filename, "w")
T = ncread(filename, "T")
P = ncread(filename, "P")


for i in 1:nb_ranks-1

    filename = string(filepath, i, ".nc")
    
    u = cat(u, ncread(filename, "u"), dims=1)
    v = cat(v, ncread(filename, "v"), dims=1)
    w = cat(w, ncread(filename, "w"), dims=1)
    T = cat(T, ncread(filename, "T"), dims=1)
    P = cat(P, ncread(filename, "P"), dims=1)

end
```

```julia
@info ncinfo(filename)
```

```julia
function to_bar(x)
    dims = size(x)
    tmp = sum(x, dims=(1, 2)) / (4 * dims[1] * dims[2])
    return repeat(tmp, outer = (dims[1], dims[2], 1, 1))

end
```

```julia
size(w)
```

```julia
u_bar = to_bar(u)
v_bar = to_bar(v)
w_bar = to_bar(w)
T_bar = to_bar(T)
P_bar = to_bar(P)

2
```

```julia
P_bar[1, 1, :, 1] == P_bar[1, 1, :, 2]
```

```julia

u_prime = u - u_bar
v_prime = v - v_bar
w_prime = w - w_bar
T_prime = T - T_bar
P_prime = P - P_bar

2
```
