```python
# Importing libraries

import numpy as np
from matplotlib import pyplot as plt
```

```python
# Some constants

time_file_path = "../results/logs/time.txt"
```

```python
# Oppening file and get the data

infile = open(time_file_path, "r")

headers = ["nb_cores", "size", "duration", "mpi", "grid", "model", "csim", "rsim"]
raw_data = [[float(x) for x in line.split("\t") if x != ""] for line in infile.read().split("\n") if line != ""]

size = len(raw_data)


t_data = {head : [line[i] for line in raw_data] for i, head in enumerate(headers)}
```

```python
# Getting the average of data

avg_data = {head : [] for head in headers}

to_combine = []
last_data = (t_data["nb_cores"][0], t_data["size"][0], t_data["duration"][0])

for i in range(size):
    data = (t_data["nb_cores"][i], t_data["size"][i], t_data["duration"][i])

    if last_data == data:
        to_combine.append(i)
        continue

    for head in headers:
        avg_data[head].append(np.mean([t_data[head][j] for j in to_combine]))

    last_data = data
    to_combine = [i]


for head in headers:
    avg_data[head].append(np.mean([t_data[head][j] for j in to_combine]))
```

```python
def filter_data(data, key, val):
    headers = data.keys()
    size = len(data[key])
    res = {head: [] for head in headers}

    for i in range(size):
        if data[key][i] != val:
            continue
            
        for head in headers:
            res[head].append(data[head][i])

    return res
```

```python
def plot_filtered(data, key1, key2, filter_args = []):
    for key, val in filter_args:
        data = filter_data(data, key, val)

    plt.plot(data[key1], data[key2])
    plt.show()

    return
```

```python
plot_filtered(avg_data, "size", "rsim", [("duration", 120)])
```

```python
plot_filtered(avg_data, "duration", "rsim", [("size", 96)])
```
